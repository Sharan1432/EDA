```python
#11.Identify and remove outliers in the km_driven column using the IQR method

import numpy as np

file_path = r'C:\Users\swamy\Downloads\BIKE DETAILS (1).csv'
df = pd.read_csv(file_path)

Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['km_driven'] < lower_bound) | (df['km_driven'] > upper_bound)]

df_clean = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

print("Number of outliers:", len(outliers))
print("Number of rows after removing outliers:", len(df_clean))

```

```python
#3.What is the distribution of discounted prices vs. actual prices

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r'C:\Users\swamy\Downloads\amazon.csv')

# Convert the actual_price and discounted_price columns to numeric
df['actual_price'] = pd.to_numeric(df['actual_price'].str.replace('$', '').str.replace(',', ''))
df['discounted_price'] = pd.to_numeric(df['discounted_price'].str.replace('$', '').str.replace(',', ''))

# Calculate the discount percentage
df['discount_percentage'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100

# Plot the histogram of discounted prices
plt.figure(figsize=(10, 6))
sns.histplot(df['discounted_price'], bins=50, kde=True)
plt.title('Distribution of Discounted Prices')
plt.xlabel('Discounted Price')
plt.ylabel('Frequency')
plt.show()

# Plot the histogram of actual prices
plt.figure(figsize=(10, 6))
sns.histplot(df['actual_price'], bins=50, kde=True)
plt.title('Distribution of Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Frequency')
plt.show()

# Plot the scatter plot of discounted prices vs. actual prices
plt.figure(figsize=(10, 6))
sns.scatterplot(x='actual_price', y='discounted_price', data=df)
plt.title('Discounted Prices vs. Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.show()

# Plot the histogram of discount percentages
plt.figure(figsize=(10, 6))
sns.histplot(df['discount_percentage'], bins=50, kde=True)
plt.title('Distribution of Discount Percentages')
plt.xlabel('Discount Percentage')
plt.ylabel('Frequency')
plt.show()
```

```python
#4.How does the average discount percentage vary across categories

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r'C:\Users\swamy\Downloads\amazon.csv')

# Convert the actual_price and discounted_price columns to numeric
df['actual_price'] = pd.to_numeric(df['actual_price'].str.replace('$', '').str.replace(',', ''))
df['discounted_price'] = pd.to_numeric(df['discounted_price'].str.replace('$', '').str.replace(',', ''))

# Calculate the discount percentage
df['discount_percentage'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100

# Group by category and calculate the average discount percentage
average_discount_by_category = df.groupby('category')['discount_percentage'].mean().reset_index()

# Sort the categories by average discount percentage in descending order
average_discount_by_category = average_discount_by_category.sort_values(by='discount_percentage', ascending=False)

# Print the result
print(average_discount_by_category)

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='category', y='discount_percentage', data=average_discount_by_category)
plt.title('Average Discount Percentage by Category')
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')
plt.xticks(rotation=90)
plt.show()
```

```python
#6.What are the most popular product keywords

import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('stopwords')

# Load the dataset
df = pd.read_csv(r'C:\Users\swamy\Downloads\amazon.csv')

# Combine product_name and description into a single column
df['product_keywords'] = df['product_name'] + ' ' + df['description']

# Convert to lowercase and tokenize
df['product_keywords'] = df['product_keywords'].apply(lambda x: x.lower())
df['product_keywords'] = df['product_keywords'].apply(word_tokenize)

# Remove stopwords
stop_words = set(stopwords.words('english'))
df['product_keywords'] = df['product_keywords'].apply(lambda x: [word for word in x if word not in stop_words])

# Join keywords back into a string
df['product_keywords'] = df['product_keywords'].apply(lambda x: ' '.join(x))

# Create a wordcloud
wordcloud = WordCloud(width=800, height=400).generate(' '.join(df['product_keywords']))

# Plot the wordcloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Find the top 10 most frequent keywords
from collections import Counter
keywords = ' '.join(df['product_keywords']).split()
keyword_counts = Counter(keywords)
print(keyword_counts.most_common(10))
```

```python
#EDA 4

#1.Read the dataframe, check null value if present then do the needful, check duplicate row , if present then do the needful

import pandas as pd

# Read the dataframe
df = pd.read_csv(r"C:\Users\swamy\Downloads\spotify.csv")

# Check for null values
null_values = df.isnull().sum()
print("Null Values:")
print(null_values)

# Handle null values (replace with mean, median, or mode)
# For this example, we'll replace with the mean
df['Popularity'] = df['Popularity'].fillna(df['Popularity'].mean())
df['Duration (ms)'] = df['Duration (ms)'].fillna(df['Duration (ms)'].mean())

# Check for duplicate rows
duplicate_rows = df.duplicated().sum()
print("\nDuplicate Rows:", duplicate_rows)

# Handle duplicate rows (drop duplicates)
df.drop_duplicates(inplace=True)

print("\nDataframe Shape after handling null values and duplicates:", df.shape)
```

```python
#6.Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist

import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\swamy\Downloads\spotify.csv")

# Get the top 5 most popular artists
top_artists = df['Artist'].value_counts().nlargest(5).index

# Calculate the average popularity for each artist
avg_popularity = df[df['Artist'].isin(top_artists)].groupby('Artist')['Popularity'].mean().reset_index()

# Print the average popularity for each artist
print("Average Popularity for each Artist:")
print(avg_popularity)

# Find the artist with the highest average popularity
max_avg_popularity_artist = avg_popularity.loc[avg_popularity['Popularity'].idxmax()]['Artist']
print(f"\nArtist with the highest average popularity: {max_avg_popularity_artist}")

```

```python
#9.Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\swamy\Downloads\spotify.csv")

# Create a box plot
plt.figure(figsize=(10,6))
plt.boxplot([df[df['Artist'] == artist]['Duration (ms)'] for artist in df['Artist'].unique()], labels=df['Artist'].unique())
plt.title('Duration of Tracks across Different Artists')
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.xticks(rotation=90)
plt.show()
```
